```python
import pandas as pd
import matplotlib.pyplot as plt

#Importo le librerie pandas e matplotlib
```

```python
df = pd.read_csv("california_housing_data.csv")
df.head()

#leggo il file csv
#e stampo le prime 5 righe 
```

```python
mean_values = df[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms','Population', 'AveOccup', 'MedHouseVal']].mean()
mean_values


#calcola la media dei valori di ogni colonna scelta
```

```python
categories = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms',
              'Population', 'AveOccup', 'MedHouseVal']

values = [mean_values[col] for col in categories]

#definisco le variabili categories e values
```

```python
plt.figure(figsize=(10,4))  
plt.bar(categories, values, color="green", edgecolor="black") #crea il grafico a barre con le categoria su asse x e valori su asse y , e assegna un colore alla barra e un colore al bordo della barra
plt.title("Grafico a barre - Media colonne California Housing Data") #titolo del grafico
plt.xlabel("Categorie") #assegna nome per l'asse x
plt.ylabel("Valori medi") #assegna nome per asse y 
plt.xticks(rotation=45) #ruota di 45 gradi le categorie asse x per avere una leggibilit√† migliore
plt.show() #Visualizza il grafico
```
